```python
import flirimageextractor
import matplotlib.pyplot as plt
from skimage.io import imread
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
# Process exported image
flir = flirimageextractor.FlirImageExtractor(palettes=[cm.jet, cm.bwr, cm.gist_ncar], is_debug=True)

flir.process_image('img_1.jpg')
thermal = flir.get_thermal_np()
rgb = flir.extract_embedded_image()
```

```python
# Plot original, thermal and RGB
fig, (ax0, ax1, ax2) =  plt.subplots(1, 3)

ax0.imshow(imread("img_1.jpg"))
ax0.set_title("Original")

ax1.imshow(thermal)
ax1.set_title("Thermal")

ax2.imshow(rgb)
ax2.set_title("RGB")
```
