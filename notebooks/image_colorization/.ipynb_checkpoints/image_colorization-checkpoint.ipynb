# Test 1 - Colorization of grayscale images

In this notebook, I am exploring the colorization of grayscale images.

The dataset in use is [GTSRB](https://www.kaggle.com/meowmeowmeowmeowmeow/gtsrb-german-traffic-sign/).
This dataset is not specifically made for this task, but it was easy to obtain and thus good to start with.

```python
import os
import numpy as np
import skimage

from skimage.io import imread
from skimage.transform import resize
from skimage.color import rgb2gray

from tensorflow.keras import Sequential
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import UpSampling2D

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,70)
```

## Load data
Currently only loading class 0.

```python
DATA_DIR = '../../data/gtsrb-german-traffic-sign/Train/0'
DIMS = (100, 100)

color, gray = [], []
for file in os.listdir(DATA_DIR):
    img = imread(os.path.join(DATA_DIR, file))
    img = resize(img, DIMS)
    color.append(img)
    gray.append(rgb2gray(img))
    
color = np.array(color)
gray = np.expand_dims(np.array(gray), axis=3)

print(gray.shape)
print(color.shape)
```

## Display example X and y

```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.imshow(color[0, ...])
ax1.title.set_text('Color')

ax2 = fig.add_subplot(122)
ax2.imshow(gray[0, ..., 0], cmap='gray')
ax2.title.set_text('Grayscale')
```

## Normalize data to range [-1, 1]

```python
X = gray * 2 - 1
y = color * 2 - 1
```

## Create and train model

When using input that has been normalized to [-1,1] (`X`), performance decreases significantly compared to using [0, 1] (`gray`).
Is this due to the relu activation function?

```python
model = Sequential()

model.add(InputLayer(input_shape=(None, None, 1)))
model.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(16, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(16, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', strides=2))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', strides=2))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(3, (3,3), activation='tanh', padding='same'))

# Finish model
model.compile(optimizer='rmsprop', loss='mse')
```

```python
# use gray, not X
model.fit(x=gray, y=y, batch_size=1, epochs=10, verbose=1)
```

## Display example output

```python
output = model.predict(gray)

fig = plt.figure()

ax1 = fig.add_subplot(131)
ax1.imshow(color[0])
ax1.title.set_text('Original color')

ax2 = fig.add_subplot(132)
ax2.imshow(gray[0, ..., 0], cmap='gray')
ax2.title.set_text('Grayscale')

ax3 = fig.add_subplot(133)
ax3.imshow((output[0] + 1) / 2, cmap='gray')
ax3.title.set_text('Colorized')
```
